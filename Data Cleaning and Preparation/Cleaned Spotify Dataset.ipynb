```python
#Importing the necessary modules
import pandas as pd
import numpy as np
```

```python
# Reading the csv file
# The parameter (encoding_errors='replace') specifies how to handle encoding errors ensuring they do not raise an exception while reading the file 
df = pd.read_csv("F:\Datasets for DS club\Most Streamed Spotify Songs 2024.csv", encoding_errors='replace')
```

```python
# Dropping Duplicates
```

```python
# ISRC is a unique and permanent identifier for each sound recording.
# Therefore, the uniqueness of a song (a row) can be determined if the ISRC is unique.
# Checking the uniqueness of the ISRC column
df['ISRC'].nunique()
```

```python
# Identify duplicate rows based on the entire row (all columns)
# Create a new dataframe with only the duplicate rows
duplicate_mask = df.duplicated(keep=False)
duplicate_rows = df[duplicate_mask]
print(duplicate_rows)
```

```python
# Dropping duplicates based on the ISRC column
df.drop_duplicates(subset=['ISRC'], inplace=True)

print("Duplicates based on the '{}' column have been dropped.".format('ISRC'))
df.info()
```

```python
# Identifying and handling missing values 
```

```python
# Checking the number of null values per column 
df.isna().sum()
```

```python
# Dropping unnecesary columns
# Specifically, drop the TIDAL Popularity column since it contains no values.
df = df.drop('TIDAL Popularity', axis=1)
```

```python
# Focus on the null records based on the Artist column.
# Specify the column to check for null values; Artist column.
column_name = 'Artist'

# Filter and display rows where the Artist column has null values.
null_records = df[column_name].isnull().sum()
print(f"Number of null_records in the column: {null_records}")
```

```python
# Displaying missing values based on the Artist column
Artist=df[df["Artist"].isnull()]
Artist
```

```python
# Dropping null records based on the Artist column
df = df.dropna(subset=['Artist'])
df.info()
```

```python
# The function name; fill_mean_round has been defined where the dataframe(df) and column_name are taken as inputs.
# It fills missing values with the mean of the column
def fill_mean_round(df, column_name):
    df[column_name] = df[column_name].fillna(df[column_name].mean()).round(0)
    return df

columns_to_fill = [
    'Apple Music Playlist Count',
    'Deezer Playlist Count',
    'Amazon Playlist Count',
    'Spotify Popularity'
]

for column in columns_to_fill:
    df = fill_mean_round(df, column)

# Check for null values
print(df.isnull().sum())
```

```python
# Converting data types
```

```python
# Convert the 'Release Date' column from an object datatype to a datetime datatype
df['Release Date'] = pd.to_datetime(df['Release Date'])
```

```python
# Modifying the 'All Time Rank' column by removing commas from the values in the column and replacing them with an empty string
df['All Time Rank'] = df['All Time Rank'].str.replace(',', '').astype(int)
```

```python
# Modifying the 'Explicit Track' column by converting the values in the column to boolean values (True or False)
df['Explicit Track'] = df['Explicit Track'].astype(bool)
```

```python
# The function name; clean_column has been defined where the dataframe(df) and column_name are taken as inputs
# It does the following:
# Replaces commas with empty strings and converts the column to float type
# Fills missing values with the mean of the column
# Converts the column into integer type
def clean_column(df, column_name):
    df[column_name] = df[column_name].str.replace(',', '').astype(float)
    mean_value = df[column_name].mean()
    df[column_name] = df[column_name].fillna(mean_value).astype(np.int64)
    return df

columns_to_clean = [
    'Spotify Streams',
    'Spotify Playlist Count',
    'Spotify Playlist Reach',
    'YouTube Views',
    'YouTube Likes',
    'YouTube Playlist Reach',
    'TikTok Posts',
    'TikTok Likes',
    'TikTok Views',
    'AirPlay Spins',
    'Deezer Playlist Reach',
    'Pandora Streams',
    'Pandora Track Stations',
    'Soundcloud Streams',
    'Shazam Counts',
    'SiriusXM Spins'
]

for column in columns_to_clean:
    df = clean_column(df, column)
```

```python
# Creating new columns
```

```python
# Creating a "Total Streams" column
df['Total Streams'] = df[['Spotify Streams','Pandora Streams','Soundcloud Streams']].sum(axis=1)
```

```python
# Creating a "Total Views" column
df['Total Views'] = df[['YouTube Views', 'TikTok Views']].sum(axis=1)
```

```python
# Getting the Radio play of each artist 
df['Radio Airplay'] = df['AirPlay Spins'] + df['SiriusXM Spins']
```

```python
# Calculating the Engagement Rate

# Engagement is a measure of how the audience interacts with the content such as through likes or comments.
# The Engagement Rate is calculated by dividing the engagement by the number of views and then multiplying by 100 to convert into a percentage.

# YouTube Engagement Rate: (YouTube Likes / YouTube Views) * 100
df['YouTube Engagement Rate'] = (df['YouTube Likes'] / df['YouTube Views']) * 100
# TikTok Engagement Rate: (TikTok Likes / TikTok Views) * 100
df['TikTok Engagement Rate'] = (df['TikTok Likes'] / df['TikTok Views']) * 100

print(df[['Track', 'YouTube Engagement Rate', 'TikTok Engagement Rate']].head())
```
